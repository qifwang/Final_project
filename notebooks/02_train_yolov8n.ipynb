```python
import os
from ultralytics import YOLO

# Notebook safe: no __file__ usage.
# The current working directory is: Final_project/notebooks/
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
DATA_DIR = os.path.join(PROJECT_ROOT, "data")
DATA_YAML = os.path.join(DATA_DIR, "survue.yaml")

print("os.getcwd():", os.getcwd())
print("PROJECT_ROOT:", PROJECT_ROOT)
print("DATA_YAML:", DATA_YAML)
```

```python
# Load a pre-trained YOLOv8n model (nano version, lightweight)
model = YOLO("yolov8n.pt")  # This will automatically download if not present.
```

```python
from ultralytics import YOLO

def on_train_epoch_start(trainer):
    print(f"\nüî• Starting Epoch {trainer.epoch + 1}/{trainer.epochs} ...")

model = YOLO("yolov8n.pt")

model.add_callback("on_train_epoch_start", on_train_epoch_start)

results = model.train(
    data=DATA_YAML,
    epochs=50,
    imgsz=640,
    batch=16,
    device="cpu",
    name="survue_yolov8n",
)
```
